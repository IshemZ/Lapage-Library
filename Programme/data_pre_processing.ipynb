```python
import biblio_modules

import pyarrow as py
import pandas as pd
import numpy as np
import time
import pytz

table_merged_PROCESSED = pd.read_csv("../Source/Data_processed/Table_merged.csv", sep=";")
```

```python
def split_column_using_space(df, new_col1, new_col2, target_col):
    df[[new_col2, new_col1]] = df[target_col].str.split(expand=True)
    return df

split_column_using_space(table_merged_PROCESSED, 'Time', 'Date', 'date')
```

```python
def convert_column_to_datetime(df, column_name):
    df[column_name] = pd.to_datetime(df[column_name],errors='coerce')
    return df
convert_column_to_datetime(table_merged_PROCESSED, 'date',)
```

```python
from datetime import datetime

def clean_data(table_merged_PROCESSED):
    table_merged_PROCESSED['Time'] = table_merged_PROCESSED['Time'].str.replace("24:", "00:", regex=False)
    table_merged_PROCESSED['Timestamp'] = table_merged_PROCESSED['Date'] + ' ' + table_merged_PROCESSED['Time']
    table_merged_PROCESSED = table_merged_PROCESSED.drop(columns=['date','Date','Time'])
    return table_merged_PROCESSED

table_merged_PROCESSED = clean_data(table_merged_PROCESSED.copy())
table_merged_PROCESSED.head()
```

```python
table_merged_PROCESSED = convert_column_to_datetime(table_merged_PROCESSED, 'Timestamp')
table_merged_PROCESSED
```

```python
table_merged_PROCESSED.to_csv('../Source/Data_processed/DateTime_cleaned.csv', sep=';', index=False)
```
